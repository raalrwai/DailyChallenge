```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customers_clean.csv", encoding = 'ISO-8859-1')
```

```python
df
```

```python
plt.hist(df['age'], bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.hist(df['age'], bins=8, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.title("Age Distribution")
plt.show()

sns.histplot(df['age'], bins=8, kde=True)
plt.title("Age Distribution with KDE")
plt.show()


sns.countplot(data=df, x='country')
plt.title("Number of Customers per Country")
plt.show()


plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='signup_date', y='age', hue='country', s=100)
plt.title("Customer Age over Signup Dates")
plt.xlabel("Signup Date")
plt.ylabel("Age")
plt.show()


plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='country', y='age')
plt.title("Age Distribution by Country")
plt.show()


plt.figure(figsize=(5,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation between Numeric Columns")
plt.show()
```
